```python
import os
import h5py
import torch
import numpy as np
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
os.chdir("..")
ROOT = os.getcwd()
```

```python
# Convert satellite images to numpy arrays
def convert_satellite():
  image_paths = [obj for obj in os.walk(ROOT + '/2020')][0][2]
  image_paths.sort()
  for image_path in image_paths:
      with h5py.File(root+'/2020/' + image_path, 'r') as f:
          data_group = f['HDFEOS']['GRIDS']['VNP_Grid_DNB']['Data Fields']
          data = np.array(data_group['Gap_Filled_DNB_BRDF-Corrected_NTL'])
          data[data > 65500] = 0

          h = image_path[18:20] 
          v = image_path[21:23] 
          np.save(root + '/2020_sat_npy/sat_h'+str(h)+'v'+str(v), data)
          del data_group, data
          f.close()
```

```python
# Convert population data to numpy arrays
def convert_population():
  root = os.getcwd() + "/drive/MyDrive/Deep Learning Project"
  population_paths = [obj for obj in os.walk(root + '/2020_population')][0][2]
  population_paths.sort()
  for count, path in enumerate(population_paths[2:]):
    pop_file = open(root + '/2020_population/' + path, 'r')
    data = pop_file.read()
    pop_file.close()

    data = data.splitlines()
    n_cols = int(data[0].split(' ')[-1])
    n_rows = int(data[1].split(' ')[-1])
    lat = int(float(data[2].split(' ')[-1]))
    lon = int(float(data[3].split(' ')[-1]))
    step = float(data[4].split(' ')[-1])
    NAN = int(data[5].split(' ')[-1])
    
    pop_array = np.zeros((n_rows, n_cols))
    for i in range(n_rows):
      row = data[6+i].split(' ')
      for j in range(n_cols):
        if int(float(row[j])) == NAN:
          pass
        else:
          pop_array[i,j] = int(float(row[j]))

    del data
    np.save(root + '/2020_pop_npy/pop_'+str(count+3), pop_array)
    del pop_array
```

```python
# Combine images into 480 x 480 x 2 pixel segments

def sat_to_pop_coords(h,v):
  """Accepts the h and v coordinates of the nighlight satelite image
  and returns the filename and coordinates of the corresponding population 
  dataset."""
  pop_num = (h // 9) + 4 * (v // 9) + 1
  filename = "pop_" + str(pop_num) + ".npy"
  x = h % 9 * 1200
  y = v % 9 * 1200
  return filename, x, y

def interpolate(arr):
  n,m = arr.shape
  new_arr = np.zeros((n*2, m*2))
  for i in range(0,2*n,2):
    for j in range(0,2*m,2):
      new_arr[i:i+2,j:j+2] = arr[i//2, j//2]

  return new_arr

def create_dataset():
  root = os.getcwd() 
  satellite_paths = [obj for obj in os.walk(root + '/2020_sat_npy')][0][2]
  satellite_paths.sort()
  for path in satellite_paths:
    h = path[5:7] 
    v = path[8:10]
    print(h)
    print(v)
    satellite_data = np.load(root + '/2020_sat_npy/' + path)
    pop_path, x, y = sat_to_pop_coords(int(h), int(v))
    population_data = np.load(root + '/2020_pop_npy/' + pop_path)[x:x+1200, y:y+1200]
    print(np.max(satellite_data))
    print(np.max(population_data))
    print('\n')
    for i in range(5):
      for j in range(5):
        data = np.zeros((480,480,2))
        data[:,:,0] = satellite_data[i*480:(i+1)*480, j*480:(j+1)*480]
        data[:,:,1] = interpolate(population_data[i*240:(i+1)*240, j*240:(j+1)*240])
        filename = 'h' + str(h) + 'v' + str(v) + '_i' + str(i) + 'j' + str(j)
        if np.max(data[:,:,1]) != 0 and np.max(data[:,:,0]) != 0 :
          np.save(root + '/2020_labeled/' + filename, data)

create_dataset()
```
